## 큐 자료구조
- 먼저 들어 온 데이터가 먼저 나가는 형식 (선입선출)
- 큐는 입구와 출구가 모두 뚫려 있는 터널과 같은 형태

```python
from collections import deque

# 규 구현을 위해 deque 라이브러리 사용
queue = deque()


queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)

queue.popleft()
queue.append(1)
queue.append(4)
queue.popleft()

print(queue) # 먼저 들어온 순서대로 출력
queue.reverse() # 역순으로 바꾸기
print(queue) # 나중에 들어온 원소부터 출력
```

### 팩토리얼 구현

```python
# n! = 1 * 2 * 3 * 4 * ...
value = 5

def factorial(value):
    result = 1
    for i in range(1, value+1):
        result *= i
    return result


# 재귀적으로 구현한 n!
def factor(value):
    if value < 1:
        return 1
    
    return value*factor(value-1)

a = factorial(5)
print(a)
b = factor(5)
print(b)

```

## DFS (Depth-First Search)
- DFS 깊이 우선 탐색이라고 부르며 그래프에서 깊은 부분을 먼전 탐색하는 알고리즘
- 스택 자료구조 혹은 재귀 함수를 이용

```python
graph = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

visted = [False] * 9
print(visted)
```

```python
def dfs(graph, v, visited):
    visited[v] = True
    print(v, end=" ")
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)
    
dfs(graph,1,visted)
```

### BFS (Breadth-Firsh Search)
- BFS는 너비 우선 탐색이라고 부르며 그래프에서 가까운 노드부터 우선적으로 탐색
- 큐 자료구조 이요

```python
from collections import deque
```

```python
graph = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

visited = [False] * 9
print(visited)
```

```python
from collections import deque
```

```python
def bfs(graph, start, visited):
    queue = deque([start])
    visited[start] = True
    
    while queue:
        v = queue.popleft()
        print(v,end=" ")
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
bfs(graph,1, visited)
        
```

### 음료수 얼려 먹기

```python
n, m = map(int, input().split())

graph = []
for i in range(n):
    graph.append(list(map(int,input())))

print(graph)
result = 0

def dfs(x,y):
    #주어진 범위에서 벗어나는 경우 즉시 종료
    if x <= -1 or x >= n or y <= -1 or y >=m:
        return False
    
    # 현재 노드를 방문하지 않았다면
    if graph[x][y] == 0:
        graph[x][y] = 1
        dfs(x-1, y)
        dfs(x,y-1)
        dfs(x+1, y)
        dfs(x,y+1)
        return True
    return False


for i in range(n):
    for j in range(m):
        # 현재 위치에서 DFS 수행하기
        if dfs(i, j) == True:
            result += 1
print(result)
```

### 미로 탈출

```python
from collections import deque

n, m = map(int, input().split())

graph = []
for i in range(n):
    graph.append(list(map(int,input())))

print(graph)

dx = [-1,1,0,0]
dy = [0,0,-1,1]

def bfs(x, y):
    queue = deque()
    queue.append((x,y))
    
    while queue:
        x, y = queue.popleft()
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx <0 or nx >= n or ny < 0 or ny >=m:
                continue
                
            # 0 의 경우 괴물이 있음
            if graph[nx][ny] == 0:
                continue
                
            if graph[nx][ny] == 1:
                graph[nx][ny] = graph[x][y] + 1
                queue.append((nx,ny))
                
    return graph[n-1][m-1]
                


print(bfs(0,0))
```
